```python
###add column month
df["Date"]=pd.to_datetime(df["Date"],format="%d-%m-%y")
df["month"] = df["Date"].dt.month
print(df)
```

```python
###add month column name(jan,feb...)
df["month_Name"] = df["Date"].dt.month_name()
print(df)
```

```python
###short name of month like(jan,feb,...)
df["month_short"] = df["Date"].dt.strftime("%b")
print(df)
```

```python
###add Year_month
df["Year_month"] = df["Date"].dt.to_period("M")
print(df)
```

```python
###month has higher average price
n = df.groupby("month_Name")["Close"].mean()
print(n)
```

```python
###Total volume per month
s = df.groupby("month_Name")["Volume"].sum()
print(s)
```

```python
### Highest per month
a = df.groupby("month_Name")["High"].max()
print(a)
```

```python
### lowest per month
x = df.groupby("month_Name")["Low"].min()
print(x)
```

```python
## total combine everything
monthly_analysis = df.groupby("month_Name").agg({"Open":"mean","Close":"mean","High":"max","Low":"min","Volume":"sum"})
print(monthly_analysis)
```
